### spotify data analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
track = pd.read_csv(r'C:\Users\DELL\Desktop\spotify\tracks.csv')
track.head()
```

```python
# check null values

track.isnull().sum()
```

```python
track.info()
```

```python
# List Top 10 popularity songs

sorted_df = track.sort_values('popularity', ascending = True).head(10)
sorted_df
```

```python
# descriptive statistics

track.describe().T
```

```python
# Top 10 most popularity songs in 90s

most_popular = track.query('popularity > 90', inplace = False).sort_values('popularity', ascending = False)
most_popular[:10]
```

```python
# set index as Release date in column

track.set_index("release_date", inplace = True)
track.index = pd.to_datetime(track.index)
track.head()
```

```python
track[["artists"]].iloc[18]
```

```python
# Convert duration_ms to duretion in seconds

track['duration'] = track['duration_ms'].apply(lambda x: round(x/1000))
track.drop('duration_ms', inplace = True, axis = 1)
```

```python
track.head()
```

```python
# who has the most monthly listeners in spotify
```

```python
# Correlation map

corr_df = track.drop(['key','mode','explicit'],axis = 1).corr(method = "pearson")
plt.figure(figsize = (14,6))
heatmap = sns.heatmap(corr_df, annot=True, fmt=".1g", vmin = -1, vmax = 1, center = 0, cmap="inferno", linewidths = 1, linecolor = "Black")
heatmap.set_title("correlation HeatMap Between Variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation = 90)
```

```python
sample_df = track.sample(int(0.004*len(track)))
```

```python
print(len(sample_df))
```

```python
# Regrassion Loudness between Energy

plt.figure(figsize = (10,6))
sns.regplot(data = sample_df, y= "loudness", x = "energy", color = "c").set(title = "Loudness Vs Energy Correlation")
```

```python
# Regrassion popularity between acousticness

plt.figure(figsize = (10,6))
sns.regplot(data = sample_df, y= "popularity", x = "acousticness", color = "b").set(title = "Popuparity Vs Acousticness Correlation")
```

```python
track['dates'] = track.index.get_level_values('release_date')
track.dates = pd.to_datetime(track.dates)
years = track.dates.dt.year
```

```python
sns.displot(years, discrete=True, aspect =2, height = 5, kind = 'hist').set(title = "Number of songs per year")
```

```python
total_dur = track.duration
fig_dims = (18,7)
fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.barplot(x = years, y=total_dur, ax = ax, errwidth = False).set(title = "Year Vs Duration")
plt.xticks(rotation=90)
```

```python
total_dur = track.duration
sns.set_style(style = "whitegrid")
fig_dims = (10,5)
fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.lineplot(x = years, y=total_dur, ax = ax).set(title = "Year Vs Duration")
plt.xticks(rotation=60)
```

```python
df_genre = pd.read_csv(r'C:\Users\DELL\Desktop\spotify\SpotifyFeatures.csv')
df_genre.head()
```

```python
# plot the graph Color_plot genre vs duration_ms

plt.title("Duration of the Songs in Different Genre")
sns.color_palette("rocket", as_cmap = True)
sns.barplot(y = 'genre', x = 'duration_ms', data = df_genre)
plt.xlabel("Duration in milli seconds")
plt.ylabel("Genres")
```

```python
# Top 5 genre by popularity

sns.set_style(style = "darkgrid")
plt.figure(figsize = (10,5))
famous = df_genre.sort_values("popularity",ascending = False).head(10)
sns.barplot(y = 'genre',x = 'popularity', data = famous).set(title = "Top 5 Genres by Popularity")
```
